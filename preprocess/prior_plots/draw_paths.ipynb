```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import os
import imageio

_background = imageio.imread('/home/sye40/Diffuser/diffuser/utils/multiagent_mountains.png')
_background = _background
data_path = '/data/prisoner_datasets/multiagent/AStar_sponsor_map/mm'
files = os.listdir(data_path)
files = [os.path.join(data_path, f) for f in files if f.endswith(".npz")]

data_path = '/data/prisoner_datasets/multiagent/AStar_sponsor_map/no_mm'
f = os.listdir(data_path)
f = [os.path.join(data_path, f) for f in files if f.endswith(".npz")]

files = files + f
fig, ax = plt.subplots(figsize =(10, 7))
fig.set_size_inches(12, 12)
ax.imshow(_background)


n_agents = 3
for file in f[:]:
    data = np.load(file, allow_pickle=True)
    red_locs = data["red_locations"]

    for i in range(n_agents):
        plt.plot(red_locs[:,i,0] , red_locs[:,i,1], zorder=10)

# Set the y-axis limits to reverse the direction
plt.xlim([0, 2428])
plt.ylim([0, 2428])
# print(red_locs[:, 1])
# ax.scatter(red_locs[:, 1], red_locs[:, 0], c='red', s=11)
```
