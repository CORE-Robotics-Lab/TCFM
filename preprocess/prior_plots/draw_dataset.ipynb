```python
import os
import numpy as np
import imageio
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

min_x = 0
max_x = 2428
min_y = 0
max_y = 2428


_background = imageio.imread('/nethome/sye40/Diffuser/diffuser/utils/no_mountain.png')

# flip the image on horizontal axis
_background = np.flip(_background, axis=0)

plt.imshow(_background, extent = [min_x, max_x, min_y, max_y])

# for each file in the directory, plot on map
path = "/data/prisoner_datasets/singleagent/sinusoidal"
files = os.listdir(path)
for file in files[:100]:
    d = np.load(os.path.join(path, file))["red_locations"].squeeze()

    plt.plot(d[:, 0], d[:, 1], 'r-', linewidth=1.0, alpha=0.5)

plt.xlim([min_x, max_x])
plt.ylim([min_y, max_y])

# plt.axis('off')
plt.title("Sinusoidal Paths")
plt.tight_layout()

```
